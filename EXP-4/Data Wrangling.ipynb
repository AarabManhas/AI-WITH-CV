```python
import pandas as pd
```

```python
# Assign data

data = {'Name': ['Aarab', 'Anushka', 'Simon', 

 'Tony', 'Natasha', 'Riya'],

 'Age': [20, 20, 18, 19, 18, 17],

 'Gender': ['M', 'F', 'M', 'M', 'F', 'F'],

 'Marks': [90, 86, 'NaN', 74, 'NaN', 71]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#computer average

c = avg = 0

for ele in df ['Marks']:

  if str(ele).isnumeric():

    c+= 1

    avg+= ele

avg/=c
```

```python
df = df.replace(to_replace="NaN",value=avg)
```

```python
df
```

```python
# Categorize gender

df['Gender'] = df['Gender'].map({'M': 0, 'F': 1 }).astype(float)
```

```python
df
```

```python
# Filter top scoring students

df = df[df['Marks'] >= 75].copy()
```

```python
# Remove age column from filtered DataFrame

df.drop('Age', axis=1, inplace=True)
```
